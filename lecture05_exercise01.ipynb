## Task 1 — Histogram: price by room type (overlapping distributions)

**What to build:** A histogram showing price distributions for **Entire home/apt vs Private room** (exclude Shared room — too few observations) overlaid on the same chart.

**Requirements:**
- Both room types on the same chart (use `color='room_type'`)
- `barmode='overlay'` with `opacity=0.6` so both distributions are visible
- A vertical line for the median of EACH room type, differently coloured
- Insight title comparing the two distributions

> 💡 `df_cap[df_cap['room_type'].isin(['Entire home/apt','Private room'])]`


```python
# Task 1
# YOUR CODE HERE
# Import required libraries

import pandas as pd
import plotly.graph_objects as go

df = pd.read_csv("airbnb_london.csv")

cap = df["price"].quantile(0.99)
df_cap = df[df["price"] <= cap].copy()

subset = df_cap[df_cap["room_type"].isin(["Entire home/apt", "Private room"])]

med_entire  = subset[subset["room_type"] == "Entire home/apt"]["price"].median()
med_private = subset[subset["room_type"] == "Private room"]["price"].median()

BAR_COLORS    = {"Entire home/apt": "#FF5A5F", "Private room": "#00A699"}
MEDIAN_COLORS = {"Entire home/apt": "#b5302e", "Private room": "#007A6E"}

fig = go.Figure()

for room_type in ["Entire home/apt", "Private room"]:
    prices = subset[subset["room_type"] == room_type]["price"]
    median = prices.median()

    fig.add_trace(go.Histogram(
        x             = prices,
        name          = room_type,
        opacity       = 0.6,
        marker_color  = BAR_COLORS[room_type],
        xbins         = dict(size=10),
    ))

    fig.add_vline(
        x                  = median,
        line_dash          = "dash",
        line_color         = MEDIAN_COLORS[room_type],
        line_width         = 2.5,
        annotation_text    = f"<b>{room_type}</b><br>median £{median:.0f}",
        annotation_position= "top right" if room_type == "Entire home/apt" else "top left",
        annotation_font_size  = 11,
        annotation_font_color = MEDIAN_COLORS[room_type],
    )

fig.update_layout(
    barmode  = "overlay",
    title    = dict(
        text      = f"Entire home/apt median £{med_entire:.0f} vs Private room median £{med_private:.0f} — Entire homes cost ~{med_entire/med_private:.1f}× more",
        font_size = 14,
        x         = 0.5,
        xanchor   = "center",
    ),
    xaxis      = dict(title="Nightly Price (£)", range=[0, cap]),
    yaxis      = dict(title="Number of Listings"),
    legend     = dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
    template   = "plotly_white",
    height     = 480,
    margin     = dict(t=110),
)

fig.show()
```

## Task 2 — Box plot: listing activity by borough

**What to build:** A **horizontal box plot** comparing listing activity (reviews per month) across London boroughs — reviews per month is a proxy for how frequently a listing is booked.

**Requirements:**
- Horizontal orientation (borough names are long)
- Sorted by median reviews per month (most active at top)
- Highlight the **two most active** boroughs in a different colour
- Outliers shown as individual points
- Insight title naming the two busiest boroughs

> 💡 Some listings have zero reviews — these are new or inactive listings. Filter them out with before plotting

```python
# Task 2
# YOUR CODE HERE

import pandas as pd
import plotly.graph_objects as go

df = pd.read_csv("airbnb_london.csv")

df_active = df[df["reviews_per_month"] > 0].copy()

medians = (
    df_active.groupby("neighbourhood")["reviews_per_month"]
    .median()
    .sort_values(ascending=True)
)
sorted_hoods = medians.index.tolist()

top2 = medians.nlargest(2).index.tolist()

HIGHLIGHT = "#FF5A5F"
DEFAULT   = "#B0C4C4"

fig = go.Figure()

for hood in sorted_hoods:
    data  = df_active[df_active["neighbourhood"] == hood]["reviews_per_month"]
    color = HIGHLIGHT if hood in top2 else DEFAULT

    fig.add_trace(go.Box(
        x            = data,
        name         = hood,
        orientation  = "h",
        marker_color = color,
        line_color   = color,
        boxpoints    = "outliers",
        marker       = dict(size=3, opacity=0.5),
    ))

fig.update_layout(
    barmode = "overlay",
    title   = dict(
        text      = f"<b>{top2[-1]}</b> and <b>{top2[-2]}</b> are London's most-booked boroughs by reviews per month",
        font_size = 14,
        x         = 0.5,
        xanchor   = "center",
    ),
    xaxis      = dict(title="Reviews per Month"),
    yaxis      = dict(title="", tickfont=dict(size=11)),
    showlegend = False,
    template   = "plotly_white",
    height     = 540,
    margin     = dict(t=90, l=200, r=40, b=60),
)

fig.show()
```
